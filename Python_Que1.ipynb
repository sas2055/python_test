## Split the Telephone Numbers into separate columns

```python
telephone_columns = df['Telephone Numbers'].str.split(',', expand=True)
telephone_columns.columns = [f"Telephone{i+1}" for i in range(telephone_columns.shape[1])]
```

```python
telephone_columns
```

## Concatenate the new telephone columns with the original DataFrame

```python
df = pd.concat([df, telephone_columns], axis=1)
df.head()
```
